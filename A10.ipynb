```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

```python
from sklearn import datasets

iris = datasets.load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df["target"] = iris.target
df.head()
```

```python
print('Number of features in dataset:', df.shape[1]-1)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(7, 7))
sns.distplot(df['sepal length (cm)'], ax=axs[0, 0])
sns.distplot(df['sepal width (cm)'], ax=axs[0, 1])
sns.distplot(df['petal length (cm)'], ax=axs[1,0])
sns.distplot(df['petal width (cm)'], ax=axs[1,1])
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10,10))
sns.boxplot(df['sepal length (cm)'], ax=axs[0, 0])
sns.boxplot(df['sepal width (cm)'], ax=axs[0, 1])
sns.boxplot(df['petal length (cm)'], ax=axs[1,0])
sns.boxplot(df['petal width (cm)'], ax=axs[1,1])
```
