```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('iris.csv')
data.head()
```

### Subtask 1:

Mean:

```python
data.groupby(by=data['Species']).mean()
```

Median:

```python
data.groupby(by=data['Species']).median()
```

Minimum:

```python
data.groupby(by=data['Species']).min()
```

Maximum:

```python
data.groupby(by=data['Species']).max()
```

Standard Deviation:

```python
data.groupby(by=data['Species']).std()
```
